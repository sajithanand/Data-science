```python
#python program for simple calculator 

def add(x,y):
    return x+y



def substract(x,y):
    return x-y


def multiply(x,y):
    return x*y


def divide(x,y):
    return x/y



print("Select option")
print("1.ADD")
print("2.Substract")
print("3.Multiplication")
print("4.Division")

while True:
    choice=input("enter choice(1/2/3/4):")
    
    if choice in('1','2','3','4'):
        num1=int(input("enter first number:"))
        num2=int(input("enter second number:"))
        
        if choice =='1':
            print(num1,"+", num2,"=",add(num1,num2))
            
        elif choice =='2':
            print("substraction result is",substract(num1,num2))
           
        elif choice =='3':
            print("multiplication result is",multiply(num1,num2))
            
        elif choice =='4':
            
            if num2==0:
                     print("inavild option")
            else:
                print("division result is",divide(num1,num2))
        
      
        else:
            print("invalid input")
            break
            
    else:
        print("exiting")
        break        
         
            
        





    
```
