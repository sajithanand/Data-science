```python
#program to draw graphs

import matplotlib.pyplot as plt
import numpy as np
a=np.array([1,2,3,4])
b=np.array([10,8,21,17])
plt.plot(a,b,marker="o")
plt.plot(a,b,mec="k")
plt.plot(a,b,ms=30)
plt.plot(a,b,color="r",linestyle="-.")

plt.show()


```

```python
#program to draw scatter graph

import matplotlib.pyplot as plt
import numpy as np
a=np.array([5,10,15,20])
b=np.array([13,7,17,19])



plt.title("scatter",loc="center")
plt.scatter(a,b)

plt.show()

```

```python
#program to draw Bar graph

import matplotlib.pyplot as plt
import numpy as np
a=np.array([1,2,3,4,5,6,7,8])
b=np.array([1,4,2,6,2,4,2,6])



plt.title("Bar graph",loc="center")
plt.bar(a,b)

plt.show()

```
