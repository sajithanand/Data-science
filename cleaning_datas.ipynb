```python
import pandas as pd
a=pd.read_csv('vt.csv')
print(a)

x=a['Calories'].mean()
print(x)


a['Calories'].fillna(x,inplace=True)
print(a)

a['Calories']=a['Calories'].astype('int')

print(a)


a['Date']=pd.to_datetime(a['Date'])
print(a)

a.duplicated().any()

a.drop_duplicates(inplace=True)
print(a)

a.duplicated().any()

```
